```r
observed <- c(254, 96)
expected <- c(263, 87)
```

```r
#Chi-Squared Goodness-of-Fit Test: takes observed values, p is used to specify the vector of expected probabilities for each category.
```

```r
result <- chisq.test(observed, p = expected / sum(expected))
```

```r
result
```

```r
# Extract the Chi-Squared test statistic from the result
```

```r
test_statistic <- result$statistic
```

```r
test_statistic
```

```r
sample_size <- 350
categories <- 2
category_probabilities <- c(0.75, 0.25)
```

```r
num_simulations <- 10000
```

```r
# Turn 10,000 # into a vector
```

```r
test_statistics <- numeric(num_simulations)
```

```r
# Do a loop. Simulated_data, will take random samples of either allele type 10,000 times. Those results are put into 'result' to be placed into chi^2 equation. result$statistic is giving me result of chi^2 test into test_statistics. The [i] next to test_statictics means it's storing the results of all 10,000 runs. 
```

```r
for (i in 1:num_simulations) {    
  simulated_data <- sample(1:categories, size = sample_size, replace = TRUE, prob = category_probabilities)
      result <- chisq.test(table(simulated_data), p = expected / sum(expected))
  test_statistics[i] <- result$statistic
}
```

```r
# To see results
```

```r
hist(test_statistics, main = "Distribution of Chi-Squared Test Statistics")
```

```r
# Results from part one saved under a different name
```

```r
observed_statistic <- test_statistic
```

```r
# Empirical pvalue. Not sure if I should be using sum or mean.
```

```r
empirical_p_value <- mean(test_statistics >= observed_statistic)
```

```r
empirical_p_value
```

# This high empirical p value shows that the null hypothesis cannot be rejected. Meaning, there is a very low probability of encountering a deviation from the anticipated outcome of the 3:1 ratio.
